```python
from requests import get
from bs4 import BeautifulSoup
import csv 
from pprint import pprint
import xlwt 
from xlwt import Workbook
```

```python
file = ("data_links.csv")
```

```python
fields = [] 
rows = [] 

with open(file, 'r') as csvfile: 
    csvreader = csv.reader(csvfile) 
    fields = next(csvreader) 
    for row in csvreader: 
#         print(row[0], '\t', row[-2])
        rows.append([row[0], row[-2]]) 

# pprint(rows)
```

```python
len(rows)
```

```python
pdf_links_with_id = []
```

```python
import time
c=0
for row in rows:
    try:
        url = row[1]
        response = get(url)
#     time.sleep(1)
        html_soup = BeautifulSoup(response.text, 'html.parser')
        a_tag = html_soup.find('a', attrs = {'itemprop':'pdfLink'}) 
        pdf_link = BeautifulSoup(str(a_tag), 'html.parser').a['href']
        pdf_links_with_id.append([row[0], pdf_link])
    except:
        pass
    c=c+1
    print(c)
```

```python
pdf_links_with_id
```

```python
wb = Workbook() 
  
sheet1 = wb.add_sheet('Sheet 1') 
i = 1
sheet1.write(0, 0, 'Id')
sheet1.write(0, 1, 'Pdf link')
for link in pdf_links_with_id:
    sheet1.write(i, 0, link[0])
    sheet1.write(i, 1, link[1])
    i += 1

wb.save('pdf_links_with_id.xls') 
```
