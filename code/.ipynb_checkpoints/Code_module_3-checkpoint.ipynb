```python
c=0
output_excel=[]
for i in loaded_model:
    for j in i[1]:
        c+=1
        temp=[i[0],j[0].replace('\n',''), j[1].replace('\n','')]
        print(i[0]+' '+j[0].replace('\n','')+' '+j[1].replace('\n',''))
        output_excel.append(temp)
```

```python
pd.DataFrame(output_excel).to_excel('output.xlsx', header=['Patent Name','Figure No.','Tumor size/Survival %'], index=False)
```
