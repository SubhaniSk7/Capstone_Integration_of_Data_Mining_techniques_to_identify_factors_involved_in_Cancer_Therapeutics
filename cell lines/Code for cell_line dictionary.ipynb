```python
import glob
import PyPDF2 
from pprint import pprint
import joblib
```

<h2>NCI-60 cell lines</h2>

```python
d1 = open("NCI-60 cell lines.txt",'r').read().split('\n')
```

```python
def remove_non_alp_num(st):
    s = ''
    for char in st:
        if char in 'qwertyuiopasdfghjklzxcvbnm0987654321':
            s += char
    return s
```

```python
nci_60_lst = []
for row in d1:
    x = row.split('\t')
    x[0] = x[0].lower()
    x[0] = x[0].replace('[','')
    x[0] = x[0].replace(']','')
    x[0] = x[0].replace(' ','')
#         x[i] = remove_non_alp_num(x[i].lower())
    nci_60_lst.append(x)    
```

```python
print('NCI-60 cell lines: ')
print('No. of entries', len(nci_60_lst))
pprint(nci_60_lst)
```

<h2>CCLE_Cell_lines</h2>

```python
d2 = open("CCLE_Cell_lines_annotations.txt",'r').read().split('\n')
```

```python
ccle_lst = []
for row in d2[1:]:
    x = row.split('\t')[0].lower().split('_')
    s = ''
    for st in x[1:]:
        s += st + ' '
    ccle_lst.append([x[0], s, 'human'])
```

```python
print('CCLE_Cell_lines: ')
print('No. of entries', len(ccle_lst))
pprint(ccle_lst)
```

<h2>ATCC cell lines</h2>

```python
atcc_lst = []
for file in glob.glob("ATCC processed files/*"):
    entries = open(file,'r').read().split('\n')
    for row in entries:
        if len(row.replace(' ','')) == 0:
            continue
        x = row.split(';')
#         x[0] = remove_non_alp_num(x[0].lower())
        x[0] = x[0].lower()
        x[0] = x[0].replace('[','')
        x[0] = x[0].replace(']','')
        x[0] = x[0].replace(' ','')
        b = x[1].split(' ')
        for val in b:
            if len(val) > 2:
                x[1] = val.replace(' ','')
                break
        atcc_lst.append(x)    
```

```python
print('ATCC cell lines: ')
print('No. of entries', len(atcc_lst))
pprint(atcc_lst)
```

<h2>Combining all the above cell lines: </h2>

```python
all_cell_lines = nci_60_lst + ccle_lst + atcc_lst
```

```python
print('All cell lines: ')
print('No. of entries', len(all_cell_lines))
```

```python
pprint(all_cell_lines)
```

```python
cell_line_names = []
for cell_line in all_cell_lines:
    cell_line_names.append(cell_line[0])
```
